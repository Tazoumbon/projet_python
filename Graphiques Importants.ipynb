```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris=load_iris()

x=iris.data
y=iris.target
names=list(iris.target_names)
```

```python
print(f'x contient {x.shape[0]} exemple et {x.shape[1]} variable')
print(f'il y a  {np.unique(y).size} classe') 
```

## Graphique de classification avec plt.scatter

```python
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
#Colorions ces points suivant la nature de leur classe
plt.scatter(x[:,0],x[:,1],c=y)
plt.xlabel('Longueur du sepal')
plt.ylabel('Largeur du sepal')
plt.show()
```

```python
#alpha pour la transparence du petal
#s(size) pour la taille du petal
#Colorions ces points suivant la nature de leur classe
plt.scatter(x[:,0],x[:,1],c=y,alpha=0.5,s=x[:,2]*100)
plt.xlabel('Longueur du sepal')
plt.ylabel('Largeur du sepal')
plt.show()
```

## Modelisation d'une fonction mathematique

```python
f = lambda x,y : np.sin(x) + np.cos(x+y)
```

```python
X=np.linspace(0,5,100)
Y=np.linspace(0,5,100)

X,Y=np.meshgrid(X,Y)
Z=f(X,Y)
```

```python
ax=plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,cmap='plasma')
plt.show()
```

## Histogramme

```python
plt.hist(x[:,0],bins=50) # bins : nombre de section
plt.xlabel('Longeur du sepal')
plt.ylabel('Frequence d apparition')
plt.show()
```

```python
# On peut creer plusieurs histogramma sur une meme figure
plt.hist(x[:,0],bins=20)
plt.hist(x[:,1],bins=20)
plt.show()
```

```python
#################################################Histogramme 2D#################################################################

#  1)
plt.hist2d(x[:,0],x[:,1])
plt.xlabel('Longueur du sepal')
plt.ylabel('Lageur du sepal')
plt.colorbar() # legend des differentes couleurs de notre histogramme
plt.show()
```

```python
from scipy import misc
face=misc.face(gray=True)
#plt.imshow(face,cmap='gray')
plt.hist(face.ravel(),bins=255)
plt.show()
```

```python
f = lambda x,y : np.sin(x) + np.cos(x+y)*np.cos(x)
```

```python
X=np.linspace(0,5,100)
Y=np.linspace(0,5,100)

X,Y=np.meshgrid(X,Y)
Z=f(X,Y)
```

```python
ax=plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,cmap='plasma')
plt.show()
```

```python
plt.contour(X,Y,Z,20,colors='black')
```

```python
plt.contourf(X,Y,Z,20,cmap='RdGy')
plt.colorbar()
```
