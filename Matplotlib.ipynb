```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,2,10)
y=x**2
```

```python
plt.plot(x,y,c='black',lw=2,ls='--')
plt.show()
```

```python
plt.scatter(x,y,c='black',lw=2)
plt.show()
```

# Cycle de vie d'une figure

```python
plt.figure(figsize=(12,8))
plt.plot(x,y)
plt.show()
```

```python
plt.figure()
plt.plot(x,y,label='Quadradique')
plt.plot(x,x**3,label='Cubique')
plt.xlabel('axes des abscisses')
plt.ylabel('axes des ordonnees')
plt.title('figure 1')
plt.legend()
plt.show()
plt.savefig('figure')

plt.figure()
plt.plot(x,y)
plt.show()
```

# Subplot

```python
plt.figure()
plt.subplot(1,2,1)
plt.plot(x,y,c='blue')
plt.subplot(1,2,2)
plt.plot(x,y,c='red')
plt.show()
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(x,y,c='blue')
plt.subplot(2,1,2)
plt.plot(x,y,c='red')
plt.show()
```

# Methode oriente object

```python
fig,ax=plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
fig,ax=plt.subplots(2,1,sharex=True)
ax[0].plot(x,y)
ax[1].plot(x,np.sin(x))
plt.show()
```

# Exercice

```python
dataset={f'experience{i}':np.random.randn(100) for i  in range(4)}
```

```python
def graphique(dataset):
    n=len(dataset)
    plt.figure(figsize=(12,8))
    for k,i in zip(dataset.keys(),range(1,n+1)):
        plt.subplot(n,1,i)
        plt.plot(dataset[k])
        plt.title(k)
    plt.show()
    
```

```python
graphique(dataset)
```
