```python
from scipy import misc
import matplotlib.pyplot as plt
face=misc.face(gray=True)
plt.imshow(face,cmap=plt.cm.gray)
plt.show()
```

```python
a=face.shape
a
```

```python
h=face.shape[0]
w=face.shape[1] 

zom_face=face[h//4:-h//4,w//4:-w//4]
zom_face[zom_face>150]=255
plt.imshow(zom_face,cmap=plt.cm.gray)
plt.show()
```
